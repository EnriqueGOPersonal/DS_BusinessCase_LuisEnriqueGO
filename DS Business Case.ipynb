## Feature Engineering

The plan is to feature engineer the next features to  obtain a single observation for each __ID_USER__.

* genero: Unique value corresponding to previous registered gender of each __ID_USER__
* monto_sum: Sum of amounts containded in __monto__
* monto_mean: Average of amounts contained in __monto__
* monto_median: Median of amounts contained in __monto__
* num_txn_week_#: Number of transactions made on the "#" week (# goes from 1 to 5 plus the "nan" option)
* num_txn_dow_#: Number of transactions made on the "#" day of week (# goes from 0 to 6 plus the "nan" option)
* establecimiento_nunique: Number of unique values for __establecimiento__
* establecimiento_X_percent: Percentage of total user's transactions with "X" __establecimiento__
* ciudad_nunique: Number of unique values for __ciudad__
* num_txn_ciudad_X: Number of transactions made for each "X" value of __ciudad__ (five values total including the "nan" option)
* tipo_tc_nunique: Number of unique values contained in __tipo_tc__ 
* num_txn_tipo_tc_X: Number of transactions made for each "X" value of __tipo_tc__ (three values total including the "nan" option)
* linea_tc_avg: Mean of amounts containded in __linea_tc__
* linea_tc_median: Median of amounts containded in __linea_tc__
* interes_tc_mean: Mean of amounts contained in __interes_tc__
* interes_tc_median: Median of amounts contained in __interes_tc__
* status_txn_X_percent: Percentage of total user's transactions with "X" __status_txn__.
* dcto_sum: Sum of amounts contained in __dcto__
* dcto_mean: Mean of amounts contained in __dcto__
* dcto_median: Sum of amounts contained in __dcto__
* cashback_sum: Sum of amounts contained in __cashback__
* cashback_mean: Mean of amounts contained in __cashback__
* cashback_median: Median of amounts contained in __cashback__
* num_transactions: Number of transactions performed by each __ID_USER__

* _Note:_ the resulting dimentions and inspection of values contained on each feature previous to the feature engineering was made using the "transactional_EDA" document available in the main repository folder.

```python
dataframe["ID_USER_copy"] = dataframe['ID_USER']
gender_df = dataframe[['ID_USER', 'genero',]].groupby("ID_USER", as_index = False).max()
sum_df = pd.get_dummies(dataframe[['ID_USER', 'fecha_week', 'fecha_dayofweek', 'ciudad', 'tipo_tc', "status_txn", "establecimiento"]],
               columns = ['fecha_week', 'fecha_dayofweek', 'ciudad', 'tipo_tc', "status_txn", "establecimiento"],
              dummy_na = True).groupby("ID_USER", as_index = False).sum()

mean_median_df = dataframe.groupby("ID_USER", as_index = False).agg({'monto': ['sum', 'mean', 'median'],
                                                   'establecimiento': 'nunique',
                                                   'ciudad': 'nunique', 
                                                   'tipo_tc': 'nunique',
                                                   'linea_tc': ['mean', 'median'],
                                                   'interes_tc': ['mean', 'median'],
                                                   'dcto': ['sum', 'mean', 'median'],
                                                    'ID_USER_copy': "count",
                                                   'cashback': ['sum', 'mean', 'median'],
                                                  })
mean_median_df.columns = ["ID_USER"] + ['{}_{}'.format(x[0], x[1]) for x in mean_median_df.columns if x[0] != "ID_USER"]
mean_median_df.columns = mean_median_df.columns.to_flat_index()
label_df = dataframe[['ID_USER', 'fraude']].groupby("ID_USER", as_index = False).max()
```

```python
grouped_dataframe = sum_df.merge(mean_median_df, on = ['ID_USER'],  how = "inner").\
merge(label_df, on = ['ID_USER'],  how = "inner").\
merge(gender_df, on = ['ID_USER'],  how = "inner")
```

```python
status_txn_cols = [c for c in grouped_dataframe if "status_txn" in c]
for c in status_txn_cols:
    grouped_dataframe[c] = grouped_dataframe[c]*100/grouped_dataframe["ID_USER_copy_count"]

establecimiento_cols = [c for c in grouped_dataframe if "establecimiento" in c and c != "establecimiento_nunique"]
for c in establecimiento_cols:
    grouped_dataframe[c] = grouped_dataframe[c]*100/grouped_dataframe["ID_USER_copy_count"] 
```

## Modelling

### Model selection

I chose to develop a Logistic Regression to model our target label. 
The key advantages of it are:
1. Easy to explain and interpret.
2. It's low computational cost allows to quickly experiment.
3. Low memory usage when saved allows it's deployment in memory restricted environments.
4. Provides a baseline of model performance for comparison with more complex models.
5. It's low computational cost allows the usage of tools with higher demand of computational resources like Cross Validation and step-forward-feature-selection, which I will later use.
6. Not particularly data demanding, allowing to obtain good performances even with small datasets.

The trade-off of using a Logistic Regression in this case are:
1. More complex models may be able to perform better.
2. Has no built-in feature engineering mechanism in its algorithm.
3. High explainability might lead to model information leakage. This point is particularly important when dealing with sensitive data or exploitable weaknesses like undesired data manipulation to obtain lower/higher scores.
4. Baseline given by LR model might not even beat human performance.

### Model building

Since we will be using cross validation we are only going to splitting the dataset into train (80% rows) and validation (20% rows) sets.

```python
label = "fraude"
id_col = "ID_USER"
```

```python
train, val = train_test_split(grouped_dataframe, test_size = 0.2, shuffle=True)
```

Building pipeline

```python
def get_num_cols(df, label, id_col):
    df = df.drop([label, id_col], axis = 1)
    num_cols = df.columns[df.dtypes.apply(pd.api.types.is_numeric_dtype)]
    return num_cols

def get_cat_cols(df, label, id_col):
    df = df.drop([label, id_col], axis = 1)
    num_cols = df.columns[df.dtypes.apply(pd.api.types.is_object_dtype)]
    return num_cols

num_cols = get_num_cols(train, label, id_col)
cat_cols = get_cat_cols(train, label, id_col)
```

```python
num_preprocessor = ColumnTransformer([
    ('numeric_imputer', SimpleImputer(strategy='median'), make_column_selector(dtype_exclude=object)),
    ('numeric_scaler', StandardScaler(), make_column_selector(dtype_exclude=object))])
cat_preprocessor = Pipeline([
    ('cat_imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(sparse = False, handle_unknown = "ignore", dtype = float))
])

param_grid_lr = {
    'classifier__C': [0.05, 0.1, 0.25, 0.5, 0.6, 0.7, 1.0],
}

estimators = [('num_prep', num_preprocessor, num_cols),
              ('cat_prep', cat_preprocessor, cat_cols)]
preprocessor = ColumnTransformer(estimators)

pipe = Pipeline([('preprocessor', preprocessor), ("classifier", LogisticRegression(max_iter = 500))])

grid_search_rf = GridSearchCV(pipe, param_grid_lr, cv = 3, n_jobs = -1, scoring ='roc_auc')

y_train = train["fraude"]
x_train = train.drop(["ID_USER", "fraude"], axis = 1)
```

Train model

```python
model = grid_search_rf.fit(x_train, y_train)
```

Selecting best parameters

```python
print("Best training weighted roc_auc was", model.best_score_, " with parameters ", model.best_params_)
```

Saving Model.

## Measuring and exploring model performance

Model's best parameters were selected by averaging the best Area under ROC score across 3 folds of the training set. Using Area under ROC curve for evaluating a model's general performance is okay as a first step, but further inspection must be made.

Another metric that I suggest for comparing models' performance is weighted F1.

Advantages of using ROC AUC over weighted F1 score are:
* Considers model performance across all probability thresholds.
* Not biased when comparing it across models coming from same sample distribution.
* Easy to plot and visually inspect.

Disadvantages of using ROC AUC over weighted F1 score are:
* ROC AUC is not a riable metric for comparing models coming from a different sample distribution.
* Might vary very little when comparing highly unbalanced models (models for which the target variable appears on a very low proportion of total observations or vice versa).
* Might vary very little when comparing highly unbalanced models (models for which the target variable appears on a very low proportion of total observations or vice versa).

The low variance between training and validation sets performance indicates the model is capable of generalising well.

```python
def generate_evaluation_df(prediction_df):
    
    thresholds = []
    tps = []
    fps = []
    tns = []
    fns = []
    
    for t in np.arange(0, 1, 0.025):
        t = round(t,2)
        thresholds.append(t)

        prediction_df.loc[prediction_df["prob_1"] <= t, "prediction_" + str(t)] = 0
        prediction_df.loc[prediction_df["prob_1"] > t, "prediction_" + str(t)] = 1

        tp = len(prediction_df.loc[(prediction_df["prediction_" + str(t)] == 1) & (prediction_df["label"] == 1)])
        fp = len(prediction_df.loc[(prediction_df["prediction_" + str(t)] == 1) & (prediction_df["label"] == 0)])
        tn = len(prediction_df.loc[(prediction_df["prediction_" + str(t)] == 0) & (prediction_df["label"] == 0)])
        fn = len(prediction_df.loc[(prediction_df["prediction_" + str(t)] == 0) & (prediction_df["label"] == 1)])

        tps.append(tp)
        fps.append(fp)
        tns.append(tn)
        fns.append(fn)
    
    evaluation_df = pd.DataFrame({"threshold": thresholds, "tp": tps, "tn": tns, "fp": fps, "fn": fns})
#     return evaluation_df
    evaluation_df["precision"] = evaluation_df.apply(lambda x: x.tp/(x.tp + x.fp + 0.00000001), axis = 1)
    evaluation_df["recall"] = evaluation_df.apply(lambda x: x.tp/(x.tp+x.fn), axis = 1)
    evaluation_df["specificity"] = evaluation_df.apply(lambda x: x.tn/(x.tn+x.fp), axis = 1)
    evaluation_df["f1s"] = evaluation_df.apply(lambda x: (x.tp*2)/(2*x.tp+x.fn+x.fp), axis = 1)
    evaluation_df["npv"] = evaluation_df.apply(lambda x: (x.tn)/(x.tn+x.fn+0.0000000001), axis = 1)

    return evaluation_df
    
evaluation_df = generate_evaluation_df(prediction_df)
```

```python
evaluation_df
```

Inspecting different probability thresholds' performance I would suggest the one with the highest F1 score (the best precision to recall ratio), that is, a threshold of 0.20

By selecting a probability threshold of 0.20, approximately 57.85% of fraudulent customers can be detected at the cost of having a false positive ratio of 66.39% (that is,  66.39% of the times one "detection" happens, it is a false alarm). This past comment might sound alarming, but looking at it from a different point of view, 75.75% of non-fraudulent customers won't any generate alert.

Depending on the seriousness of actions derived from a false positive classification, we might be more cautious and select an upper threshold to meet our expectations. For example, with a 0.62 threshold we would only wrongly classify 1% of non-fraudulent customers.

The suggested approach in the end is to generate a series of actions to take based on the confidence of the prediction. While a prediction with a low probability of fraud might only require some "caution" flag, a prediction with 99% probability may result in an instant account-block.

```python
def generate_metric_graph(evaluation_df, title = "Desempeño de métricas en función de Threshold"):
    
    temp_df = pd.DataFrame(evaluation_df[["recall", "threshold"]].values, columns = ["value", "threshold"])
    temp_df["metric"] = "Recall"
    temp_df_1 = pd.DataFrame(evaluation_df[["precision", "threshold"]].values, columns = ["value", "threshold"])
    temp_df_1["metric"] = "Precision"
    temp_df_2 = pd.DataFrame(evaluation_df[["specificity", "threshold"]].values, columns = ["value", "threshold"])
    temp_df_2["metric"] = "Specificity"
    temp_df_3 = pd.DataFrame(evaluation_df[["npv", "threshold"]].values, columns = ["value", "threshold"])
    temp_df_3["metric"] = "NPV"
    temp_df = temp_df.append(temp_df_1)
    temp_df_2 = temp_df_2.append(temp_df_3)

    a4_dims = (30, 20)
    fig, axis = plt.subplots(1, 2, figsize=a4_dims)
    fig.suptitle(title, size = 15)

    a = sns.lineplot(x = "threshold", y = "value", data=temp_df, hue = "metric", palette=["skyblue", "coral"], ax = axis[0])
    a.set_title("Precision vs Recall", size = 15)
    a.set_xlabel("Threshold")
    a.set_ylabel("Value")
    a.set_xticks(np.arange(0.0,1.05,0.05))
    a.set_yticks(np.arange(0.0,1.05,0.05))
    
    b = sns.lineplot(x = "threshold", y = "value", data=temp_df_2, hue = "metric", palette=["skyblue", "coral"], ax = axis[1])
    b.set_title("Specificity vs NPV", size = 15)
    b.set_xlabel("Threshold")
    b.set_ylabel("Value")
    b.set_xticks(np.arange(0.0,1.05,0.05))
    b.set_yticks(np.arange(0.0,1.05,0.05))
    print(a)
    
    return 

generate_metric_graph(evaluation_df, "Metrics performance by Threshold \n")
```
